```python
categoriesSaarland = {
    'covid' : ['corona', 'lauterbach', 'corona-maßnahmen', 'rki', 'inzidenz', 'corona-virus',
               'robert-koch-institut', 'infektionen', 'corona-inzidenz', 'neuinfektionen',
               'robert-koch-instituts', 'pandemie'],
    
    'unglück' : ['feuerwehr', 'unfall', 'brand', 'einsatz', 'feuer', 'gesperrt', 'unglück'],
    
    'energiekrise' : ['preise', 'gas', 'energiekrise', 'energie', 'strom', 'sparen', 'spritpreise'],
    
    'wetterKlima' : ['wetter', 'sommer', 'dwd', 'temperaturen', 'wetterdienst', 'gewitter', 'hitze', 'sonne', 'regen', 'grad'],
    
    'ukraine' : ['ukraine', 'krieg', 'russland', 'russischen', 'ukraine-krieg', 'ukrainische', 'kiew', 'putin', 'ukrainischen', 'russische', 'russlands', 'moskau'],
    
    'politik' : ['spd', 'cdu', 'landtagswahl', 'landtag', 'rehlinger', 'hans', 'ministerpräsidentin',
                 'grünen', 'partei', 'afd', 'fdp', 'wahl', 'ministerpräsident', 'amt', 'linke', 'grüne'],
    
    'polizistenmordKusel' : ['kusel'],
    
    'lokalesSaarbrücken' : ['saarbrücken', 'saarbrücker', 'landeshauptstadt', 'nauwieser', 'püttlingen']
}
```

```python
dfsSaarland = []

for key,values in categoriesSaarland.items():
    words=pd.DataFrame({'category': key, 'term': values})
    dfsSaarland.append(words)

termsSaarland = pd.concat(dfsSaarland)
termsSaarland
```

```python
# Take the 'term' column from our list of terms
termlistSaarland = list(set(termsSaarland.term))

# binary=True only does 0/1
# vocabulary= is the list of words we're interested in tracking
vectorizer = TfidfVectorizer(binary=False, vocabulary = termlistSaarland)

matrix = vectorizer.fit_transform(saarland.text, )

words_df = pd.DataFrame(matrix.toarray(), columns = vectorizer.get_feature_names_out())
```

```python
# Group the terms by category, then loop through each category
for category_name, rows in termsSaarland.groupby('category'):
    # Convert the terms for that category into a simple list
    # for example, ['student', 'educ', 'teacher']
    terms = list(rows['term'])
    print(f"Topic: {category_name} \nWorte: {terms} \n")

    # words_df[terms]  -> die einzelnen bestandteile der categories (also die wörter der topics)
    # .any(axis=1) -> prüfen, ob irgendeines von den wörtern != 0
    # .astype(int) macht aus True bzw. False einen binären atomic vector
    # df[category_name] -> wird dann zu dem jeweiligen df als variable "category_name" zugewiesen
    saarland[category_name] = words_df[terms].any(axis=1).astype(int)
```
