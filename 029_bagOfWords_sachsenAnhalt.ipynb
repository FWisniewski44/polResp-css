```python
categoriesSachsenAnhalt = {
    'unglück' : ['unfall', 'verletzt', 'gesperrt', 'stirbt', 'verletzte', 'kracht'],
    
    'wetterKlima' : ['sturm', 'wetter', 'gewitter', 'unwetterwarnung'],
    
    'großprojekte' : ['intel', 'sanierung', 'baut', 'immobilien'],
    
    'energiekrise' : ['gas', 'strom', 'energiekrise'],
    
    'justiz' : ['polizei', 'ermittelt', 'haft', 'unbekannte', 'vermisst', 'zeugen', 'täter'],
    
    'covid' : ['inzidenz', 'rki', 'neuinfektionen'],
    
    'ukraine' : ['ukraine', 'russland', 'ukraine-krieg', 'krieg', 'putin', 'ukrainische', 'russische',
                 'kiew', 'flüchtlinge', 'russischer', 'angriff', 'sanktionen', 'waffen', 'russischen',
                 'russlands', 'mariupol', 'selenskyj']
}
```

```python
dfsSachsenAnhalt = []

for key,values in categoriesSachsenAnhalt.items():
    words=pd.DataFrame({'category': key, 'term': values})
    dfsSachsenAnhalt.append(words)

termsSachsenAnhalt = pd.concat(dfsSachsenAnhalt)
termsSachsenAnhalt
```

```python
# Take the 'term' column from our list of terms
termlistSachsenAnhalt = list(set(termsSachsenAnhalt.term))

# binary=True only does 0/1
# vocabulary= is the list of words we're interested in tracking
vectorizer = TfidfVectorizer(binary=False, vocabulary = termlistSachsenAnhalt)

matrix = vectorizer.fit_transform(sachsenAnhalt.text, )

words_df = pd.DataFrame(matrix.toarray(), columns = vectorizer.get_feature_names_out())
```

```python
# Group the terms by category, then loop through each category
for category_name, rows in termsSachsenAnhalt.groupby('category'):
    # Convert the terms for that category into a simple list
    # for example, ['student', 'educ', 'teacher']
    terms = list(rows['term'])
    print(f"Topic: {category_name} \nWorte: {terms} \n")

    # words_df[terms]  -> die einzelnen bestandteile der categories (also die wörter der topics)
    # .any(axis=1) -> prüfen, ob irgendeines von den wörtern != 0
    # .astype(int) macht aus True bzw. False einen binären atomic vector
    # df[category_name] -> wird dann zu dem jeweiligen df als variable "category_name" zugewiesen
    sachsenAnhalt[category_name] = words_df[terms].any(axis=1).astype(int)
```
