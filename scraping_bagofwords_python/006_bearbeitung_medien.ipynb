## Überregionale Mediendaten

```python
# laden der mediendatensätze überregional

bisPhoenix = pd.read_csv("/home/fw/Documents/uni/masterarbeit/scraping/polResp-css/sicherung_medien_bisPhoenix.csv", encoding="UTF-8", index_col=0)
bisEnde = pd.read_csv("/home/fw/Documents/uni/masterarbeit/scraping/polResp-css/sicherung_medien_abReuters.csv", encoding="UTF-8", index_col=0)
bisEnde
```
