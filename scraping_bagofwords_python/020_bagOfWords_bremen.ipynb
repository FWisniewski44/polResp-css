```python
categoriesBremen = {
    'lokalesBremen' : ['bremens', 'bremen', 'bürgermeister', 'bovenschulte', 'hemelingen', 'borgfeld',
                       'rathaus', 'bremer', 'innenstadt', 'freimarkt', 'bremerin', 'bremerinnen',
                       'verden', 'vegesack', 'breminale'],
    
    'justiz' : ['verletzt', 'zeugen', 'täter', 'unfall', 'hinweise', 'ermittelt', 'festgenommen'],
    
    'ukraine' : ['ukraine', 'russland', 'krieg', 'russischen', 'putin', 'russische', 'russlands',
                 'kiew', 'ukrainische', 'moskau', 'sanktionen', 'selenskyj', 'angriff', 'ukrainischen'],
    
    'politik' : ['spd', 'grünen', 'cdu', 'bürgerschaft', 'senat', 'fdp', 'partei', 'wahl',
                 'linken', 'bürgerschaftswahl'],
    
    'energiekrise' : ['gas', 'kosten', 'preise', 'energie', 'swb', 'verbraucher', 'strom', 'sparen'],
}
```

```python
pp.pprint(categoriesBremen)
```

```python
dfsBremen = []

for key,values in categoriesBremen.items():
    words=pd.DataFrame({'category': key, 'term': values})
    dfsBremen.append(words)

termsBremen = pd.concat(dfsBremen)
termsBremen
```

```python
# Take the 'term' column from our list of terms
termlistBremen = list(set(termsBremen.term))

# binary=True only does 0/1
# vocabulary= is the list of words we're interested in tracking
vectorizer = TfidfVectorizer(binary=False, vocabulary = termlistBremen)

matrix = vectorizer.fit_transform(bremen.text, )

words_df = pd.DataFrame(matrix.toarray(), columns = vectorizer.get_feature_names_out())
```

```python
# Group the terms by category, then loop through each category
for category_name, rows in termsBremen.groupby('category'):
    # Convert the terms for that category into a simple list
    # for example, ['student', 'educ', 'teacher']
    terms = list(rows['term'])
    print(f"Topic: {category_name}\nWörter: {terms}\n")

    # words_df[terms] gets the columns for 'student', 'educ', and 'teacher'
    # .any(axis=1) sees if any of them are a 1, gives True/False
    # .astype(int) converts True/False to 1/0
    # df[category_name] = will assign that value to df['education']
    bremen[category_name] = words_df[terms].any(axis=1).astype(int)
```
