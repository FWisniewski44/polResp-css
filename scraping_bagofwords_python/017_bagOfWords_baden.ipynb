```python
categoriesBawü = {
    'unglücke' : ['unfall', 'verletzt', 'gesperrt', 'verletzte', 'krankenhaus', 'stürzt', 'verkehrsunfall'],
    
    'justiz' : ['gericht', 'prozess', 'haft', 'fall', 'urteil', 'fälle', 'verurteilt', 'vorwürfe',
                'zeugen', 'tote', 'ermittelt', 'unbekannte', 'opfer', 'täter', 'ermittlungen',
                'getötet', 'festgenommen', 'messer', 'flucht', 'fall'],
    
    'energiekrise' : ['gas', 'preise', 'steigen', 'inflation', 'energie', 'kosten', 'strom',
                      'energiekrise', 'teurer', 'kunden', 'sparen'],
    
    'kinderbetreuung' : ['kinder', 'eltern', 'kindern', 'jugendliche', 'kita'],
    
    'covid' : ['corona', 'impfpflicht', 'pandemie', 'maskenpflicht', 'omikron', 'coronavirus', 'kliniken'],
    
    'ukraine' : ['ukraine', 'russland', 'krieg', 'putin', 'kiew', 'russischen', 'russische',
                 'ukrainekrieg', 'ukrainische', 'ukraine-krieg', 'selenskyj', 'moskau',
                 'russlands', 'flüchtlinge', 'angriff', 'nato'],
    
    'öpnvVerkehr' : ['bahn', '9-euro-ticket', 'zug', 'fahren', 'bus', 'deutschebahn', '9euroticket'],
    
    'politik' : ['bürgermeister', 'cdu', 'spd', 'wahl', 'politik', 'grüne', 'afd',
                 'fdp', 'grünen', 'palmer', 'oberbürgermeister']
}
```

```python
pp.pprint(categoriesBawü)
```

```python
dfsBawü = []

for key,values in categoriesBawü.items():
    words=pd.DataFrame({'category': key, 'term': values})
    dfsBawü.append(words)

termsBawü = pd.concat(dfsBawü)
termsBawü
```

```python
# Take the 'term' column from our list of terms
termlistBawü = list(set(termsBawü.term))

# binary=True only does 0/1
# vocabulary= is the list of words we're interested in tracking
vectorizer = TfidfVectorizer(binary=False, vocabulary = termlistBawü)

matrix = vectorizer.fit_transform(bawü.text, )

words_df = pd.DataFrame(matrix.toarray(), columns = vectorizer.get_feature_names_out())
```

```python
# Group the terms by category, then loop through each category
for category_name, rows in termsBawü.groupby('category'):
    # Convert the terms for that category into a simple list
    # for example, ['student', 'educ', 'teacher']
    terms = list(rows['term'])
    print(f"Topic: {category_name}\nWörter: {terms}\n")

    # words_df[terms] gets the columns for 'student', 'educ', and 'teacher'
    # .any(axis=1) sees if any of them are a 1, gives True/False
    # .astype(int) converts True/False to 1/0
    # df[category_name] = will assign that value to df['education']
    bawü[category_name] = words_df[terms].any(axis=1).astype(int)
```
