```python
categoriesRlp = {    
    'energiekrise' : ['preise', 'kunden', 'gas', 'kosten', 'teurer', 'energiekrise', 'sparen', 'strom',
                      'energie', 'verbraucher', 'inflation', 'spritpreise'],
    
    'ukraine' : ['ukraine', 'russland', 'krieg', 'putin', 'russischen', 'russische', 'ukrainische',
                 'ukraine-krieg', 'ukrainischen', 'geflüchtete', 'russlands', 'angriff', 'kiew', 'moskau'],
    
    'unglückUnfall' : ['verletzt', 'unfall', 'gesperrt', 'verkehrsunfall'],
    
    'politik' : ['cdu', 'spd', 'oberbürgermeister', 'ebling', 'amt', 'wahl', 'gewählt', 'bürgermeister', 'grünen',
                 'kandidaten'],
    
    'covid' : ['corona', 'coronavirus', 'lauterbach', 'maskenpflicht', 'corona-regeln', 'impfung',
               'corona-maßnahmen', 'inzidenz', 'neuinfektionen', 'pandemie', 'omikron'],
    
    'unglückBrände' : ['feuerwehr', 'einsatz', 'brand', 'feuer'],
    
    'flutAhrtal' : ['ahrtal', 'flutkatastrophe', 'flut', 'ahrweiler', 'ahr'],
    
    'polizistenmordKusel' : ['kusel', 'polizistenmord']
}
```

```python
pp.pprint(categoriesRlp)
```

```python
dfsRlp = []

for key,values in categoriesRlp.items():
    words=pd.DataFrame({'category': key, 'term': values})
    dfsRlp.append(words)

termsRlp = pd.concat(dfsRlp)
termsRlp
```

```python
# Take the 'term' column from our list of terms
termlistRlp = list(set(termsRlp.term))

# binary=True only does 0/1
# vocabulary= is the list of words we're interested in tracking
vectorizer = TfidfVectorizer(binary=False, vocabulary = termlistRlp)

matrix = vectorizer.fit_transform(rlp.text, )

words_df = pd.DataFrame(matrix.toarray(), columns = vectorizer.get_feature_names_out())
```

```python
# Group the terms by category, then loop through each category
for category_name, rows in termsRlp.groupby('category'):
    # Convert the terms for that category into a simple list
    # for example, ['student', 'educ', 'teacher']
    terms = list(rows['term'])
    print(f"Topic: {category_name} \nWorte: {terms} \n")

    # words_df[terms]  -> die einzelnen bestandteile der categories (also die wörter der topics)
    # .any(axis=1) -> prüfen, ob irgendeines von den wörtern != 0
    # .astype(int) macht aus True bzw. False einen binären atomic vector
    # df[category_name] -> wird dann zu dem jeweiligen df als variable "category_name" zugewiesen
    rlp[category_name] = words_df[terms].any(axis=1).astype(int)
```
